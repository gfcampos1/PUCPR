## Variáveis

Em computação, uma variável é um nome simbólico associado a um valor ou a um local na memória, onde esse valor é armazenado. Variáveis permitem que os programas armazenem e manipulem dados de maneira dinâmica, fornecendo uma forma de referenciar e alterar esses dados ao longo da execução do programa.

> Características de Variáveis:
Nome: Identificador simbólico utilizado para referenciar a variável.
Tipo: Tipo de dado que a variável pode armazenar (por exemplo, inteiro, ponto flutuante, string).
Valor: Conteúdo atual armazenado na variável.
Escopo: Região do código onde a variável é válida e acessível.
Tempo de vida: Período durante o qual a variável existe na memória.

```
a = 'May'
b = 18

print(a,b)

a = 20
print(a,b)
```

## Funções e Escopos

As funções permitem modularidade, facilitando a organização e manutenção do código. O escopo, por outro lado, define a visibilidade e a duração de variáveis dentro de um programa. Existem dois tipos principais de escopo: o escopo global, onde as variáveis são acessíveis em qualquer parte do programa, e o escopo local, onde as variáveis são acessíveis apenas dentro da função ou bloco de código onde foram definidas. 

<img src="image-20240717-134748.png" width="" align="" />

```
x = 100  # Variável global

def funcao():
    y = 500  # Variável local
    print('Dentro Funcao x=', x)  # Acessando variável global dentro da função
    print('Dentro Funcao y=',y) #Acessando variável local da função

funcao()
print('Fora Funcao x=',x)
#print('Fora Funcao y=',y) # Acessando variável local fora da função (erro: out of scope/undefined variable)
```

### return

> Usamos o return para repassar (ou retornar) o valor de uma variável local de uma função para o ambiente de chamada. Isso é essencial para tornar os valores calculados dentro de uma função acessíveis fora dela.

<img src="image-20240717-135330.png" width="" align="" />

```
x = 10  # Variável global

def funcao():
    y = 5  # Variável local
    print(x)  # Acessando variável global dentro da função
    print(y)
    return y

my_var = funcao()
print(x)
print(my_var)
```

## Operadores Aritméticos

> Adição (+) ==>  10 + 3 ==> 15
Subtração (-) ==> 10 - 3 ==> 7
Multiplicação (*) ==> 10 * 3 ==> 30
Divisão (/) ==> 10/3 ==> 3.3333
Divisão Inteira (//) ==> 10//3 ==> 3
Módulo (%) ==> 10%3 ==> 1
Exponenciação (**) ==> 10**3 ==> 1000

```
a = 10
b = 3

# Adição
resultado = a + b
print(f"A adição de {a} e {b} é: {resultado}")

# Subtração
resultado = a - b
print(f"A subtração de {a} e {b} é: {resultado}")

# Multiplicação
resultado = a * b
print(f"A multiplicação de {a} e {b} é: {resultado}")

# Divisão
resultado = a / b
print(f"A divisão de {a} por {b} é: {resultado}")

# Divisão Inteira
resultado = a // b
print(f"A divisão inteira de {a} por {b} é: {resultado}")

# Módulo
resultado = a % b
print(f"O módulo de {a} por {b} é: {resultado}")

# Exponenciação
resultado = a ** b
print(f"{a} elevado à potência de {b} é: {resultado}")
```

## Operadores Lógicos (E, OU e NEGAÇÃO)

<img src="image-20240716-145830.png" width="" align="" />

```
# Operador AND (&&)
x = True
y = False
resultado = x and y
print(f"AND (x and y) ==> {x} and {y} ==> {resultado}")

# Operador OR (||)
resultado = x or y
print(f"OR (x or y) ==> {x} or {y} ==> {resultado}")

# Operador NOT
resultado = not x
print(f"NOT (not x) ==> not {x} ==> {resultado}")

# Operador AND com expressões
a = 10
b = 3
resultado = (a > 5) and (b < 5)
print(f"AND com expressões ==> (a > 5) and (b < 5) ==> {resultado}")

# Operador OR com expressões
resultado = (a < 5) or (b < 5)
print(f"OR com expressões ==> (a < 5) or (b < 5) ==> {resultado}")

# Operador NOT com expressão
resultado = not (a > 5)
print(f"NOT com expressão ==> not (a > 5) ==> {resultado}")

# XOR com Booleanos
x = True
y = False
resultado = x ^ y
print(f"XOR (x ^ y) ==> {x} ^ {y} ==> {resultado}")

x = True
y = True
resultado = x ^ y
print(f"XOR (x ^ y) ==> {x} ^ {y} ==> {resultado}")

x = False
y = False
resultado = x ^ y
print(f"XOR (x ^ y) ==> {x} ^ {y} ==> {resultado}")

# XOR com Inteiros
a = 10  # 1010 em binário
b = 3   # 0011 em binário
resultado = a ^ b
print(f"XOR (a ^ b) ==> {a} ^ {b} ==> {resultado} (binário: {bin(resultado)})")
```

## Expressões Lógicas Compostas

As expressões lógicas compostas permitem a criação de condições complexas ao combinar múltiplas expressões simples através dos operadores lógicos.

*iNSERIR PRECEDENCIAS LOGICAS

```
a = 10
b = 3
c = 7

# Expressão composta usando AND e OR
resultado = (a > 5) and (b < 5) or (c == 7)
print(f"((a > 5) and (b < 5)) or (c == 7) ==> ({a} > 5) and ({b} < 5) or ({c} == 7) ==> {resultado}")

# Outro exemplo de expressão composta com AND e OR
resultado = (a < 5) or (b > 1) and (c != 7)
print(f"((a < 5) or (b > 1)) and (c != 7) ==> ({a} < 5) or ({b} > 1) and ({c} != 7) ==> {resultado}")

# Expressão composta usando NOT, AND e OR
resultado = not (a < 5) and (b == 3) or (c > 10)
print(f"not (a < 5) and (b == 3) or (c > 10) ==> not ({a} < 5) and ({b} == 3) or ({c} > 10) ==> {resultado}")

# Outro exemplo de expressão composta com NOT, AND e OR
resultado = not (a > 5) or (b != 3) and (c < 8)
print(f"not (a > 5) or (b != 3) and (c < 8) ==> not ({a} > 5) or ({b} != 3) and ({c} < 8) ==> {resultado}")

# Expressão composta usando XOR com AND e OR
x = True
y = False

# XOR aplicado a booleanos
resultado = (a > 5) ^ (b < 5)
print(f"(a > 5) ^ (b < 5) ==> ({a} > 5) ^ ({b} < 5) ==> {resultado}")

# XOR com AND e OR
resultado = ((a > 5) and (b < 5)) ^ ((c < 10) or (a == 10))
print(f"((a > 5) and (b < 5)) ^ ((c < 10) or (a == 10)) ==> (({a} > 5) and ({b} < 5)) ^ (({c} < 10) or ({a} == 10)) ==> {resultado}")
```

## Estruturas de Seleção

As estruturas de seleção, controlam o fluxo de execução de um programa. Elas permitem que diferentes blocos de código sejam executados dependendo do resultado de uma expressão condicional. Por exemplo:

<img src="image-20240717-142448.png" width="" align="" />

```
# Solicita ao usuário que digite a nota
nota = float(input("Digite a nota do aluno: "))

# Estrutura de seleção para classificar a nota
if nota >= 90:
    print("Desempenho: Excelente")
    print("Parabens")
elif nota >= 80:
    print("Desempenho: Muito Bom")
    print("Parabens")
elif nota >= 70:
    print("Desempenho: Bom")
    print("Você pode se dedicar")
elif nota >= 60:
    print("Desempenho: Satisfatório")
else:
    print("Desempenho: Insuficiente")

print('Algoritmo finalizado.')
print("Teste")
```

```
# Solicita ao usuário que digite a idade e se é estudante
idade = int(input("Digite a sua idade: "))
estudante = input("Você é estudante? (sim/não): ").strip().lower

# Estrutura de seleção com condições compostas
if idade < 18 and estudante == "sim":
    print("Você é elegível para o desconto estudantil para menores de 18 anos.")
elif idade >= 18 and idade <= 25 and estudante == "sim":
    print("Você é elegível para o desconto estudantil para jovens adultos.")
elif idade > 60 or estudante == "sim":
    print("Você é elegível para o desconto especial para idosos ou estudantes.")
else:
    print("Você não é elegível para nenhum desconto especial.")

print('Algoritmo finalizado.')
```

## Estruturas de Repetição

Os laços de repetição permitem que blocos de código sejam executados repetidamente enquanto uma condição específica for verdadeira. Em Python, existem dois principais tipos de estruturas de repetição: while e for.

<img src="image-20240717-143010.png" width="" align="" />

A estrutura while executa um bloco de código repetidamente enquanto uma condição especificada for verdadeira.

```
print("Antes do WHILE")
i=8
while i < 100:
    print(i)
    i += 20

print("Acabou")
```

A estrutura for é usada para iterar sobre uma sequência definida (mais a frente veremos como iterar sobre estruturas de dados)

```
print("Antes do FOR")

for i in range(10):
    print(i)

print("Acabou")
```
