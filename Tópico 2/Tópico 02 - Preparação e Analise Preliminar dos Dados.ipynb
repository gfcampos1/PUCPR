```
# Contar o número de valores faltantes por coluna
missing_values_count = df.isnull().sum()

# Calcular o percentual de valores faltantes por coluna
missing_percentage = (missing_values_count / len(df)) * 100

# Criar um DataFrame para exibir os resultados
missing_data = pd.DataFrame({
    'Missing Values': missing_values_count,
    'Percentage': missing_percentage
})

# Exibir o DataFrame com os valores e percentuais de dados faltantes
print(missing_data)

# Exibir as colunas com dados faltantes (opcional)
print("\nColunas com dados faltantes:")
print(missing_data[missing_data['Missing Values'] > 0])

# Exibir o número e percentual de linhas com mais de um dado ausente
print(f"Número de linhas com mais de um dado ausente: {rows_with_multiple_missing_count}")
print(f"Percentual de linhas com mais de um dado ausente: {percentage_with_multiple_missing:.2f}%")
```

## Tratamento de Redundância

Agora que o dataset está sem dados faltantes, uma segunda etapa é a eliminação de dados redundantes. Isso pode ser feito de algumas formas:

- Visualmente com informações contextuais do problema

- Verificação de linhas repetidas (se houver)

- Matriz de correlação

```
# Identificar e contar o número de ocorrências de linhas duplicadas
duplicated_rows = df[df.duplicated(keep=False)]  # keep=False marca todas as duplicatas

# Contar o número de ocorrências de cada linha duplicada
occurrence_counts = duplicated_rows.groupby(duplicated_rows.columns.tolist()).size()

# Exibir as ocorrências de linhas duplicadas
print("Número de ocorrências de linhas duplicadas:")
print(occurrence_counts)

# Agrupar e exibir as linhas duplicadas com mais de uma n_occurences
n_occurences = 7
rows_with_multiple_occurrences = occurrence_counts[occurrence_counts > n_occurences]

print(f"\nNúmero de linhas duplicadas com mais de uma ocorrência: {n_occurences}")
print("\nLinhas duplicadas com mais de uma ocorrência:")
print(rows_with_multiple_occurrences)

# Exibir o número de linhas antes e depois da remoção
print(f"\nNúmero de linhas antes de remover duplicatas: {len(df)}")
# Remover duplicatas, mantendo a primeira ocorrência
df = df.drop_duplicates()
print(f"Número de linhas após remover duplicatas: {len(df)}")
```

Uma outra análise estatística que determina a correlação entre variáveis é a matriz de correlação.

Importante, o cálculo de correlação somente é possível entre colunas numéricas 

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionar colunas numéricas para calcular a matriz de correlação
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calcular a matriz de correlação
correlation_matrix = numeric_df.corr()

# Configurar o tamanho do gráfico
plt.figure(figsize=(12, 10))

# Criar um gráfico de calor para a matriz de correlação
ax = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Configurar o título do gráfico
plt.title('Matriz de Correlação das Colunas Numéricas')

# Colocar os rótulos do eixo x na parte superior
ax.xaxis.set_label_position('top')
ax.xaxis.set_ticks_position('none')  # Remove os ticks para melhor visualização
ax.xaxis.set_tick_params(top=True, labeltop=True)  # Mostrar os rótulos no topo

# Exibir o gráfico
plt.show()
```

E como adicionar outras variáveis categóricas

- Codificação do Label:  É uma técnica de pré-processamento usada para converter variáveis categóricas em formato numérico. Cada valor único de uma coluna categórica é atribuído a um número inteiro distinto.

- "Azul", "Vermelho", "Verde" --> [0,1,2]

```
from sklearn.preprocessing import LabelEncoder

print(df.head(5))
# Identificar colunas categóricas
categorical_columns = df.select_dtypes(include=['object','boolean']).columns

# Criar um objeto LabelEncoder
label_encoders = {}
reverse_mappings = {}
# Aplicar Label Encoding para cada coluna categórica
for column in categorical_columns:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column].astype(str))  # Convertendo para string antes de aplicar Label Encoding
    label_encoders[column] = encoder  # Armazenar o encoder para possíveis usos futuros
    
    #aqui utilizaremos um mapeamento reverso, criando pares (label original, encoded)
    reverse_mappings[column] = dict(zip(encoder.transform(encoder.classes_), encoder.classes_))

# Exibir as primeiras linhas do DataFrame transformado
print(df.head())

# Exibir as colunas transformadas
print("Colunas após Label Encoding:")
print(df.columns)
print(reverse_mappings)
```

Agora vamos visualizar a matriz de correlação, para analisar com os atributos categóricos

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionar colunas numéricas para calcular a matriz de correlação
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calcular a matriz de correlação
correlation_matrix = numeric_df.corr()

# Configurar o tamanho do gráfico
plt.figure(figsize=(20, 20))

# Criar um gráfico de calor para a matriz de correlação
ax = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Configurar o título do gráfico
plt.title('Matriz de Correlação das Colunas Numéricas')

# Colocar os rótulos do eixo x na parte superior
ax.xaxis.set_label_position('top')
ax.xaxis.set_ticks_position('none')  # Remove os ticks para melhor visualização
ax.xaxis.set_tick_params(top=True, labeltop=True)  # Mostrar os rótulos no topo

# Exibir o gráfico
plt.show()
```

Podemos agora perceber alguns atributos altamente correlacionados. Vamos reduzir-los para não impactar em determinadas análises.

```
 #Criar uma máscara para a matriz de correlação
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)

# Encontrar pares de colunas altamente correlacionadas
high_corr_var = np.where(mask & (correlation_matrix > 0.95))
high_corr_pairs = [(correlation_matrix.columns[x], correlation_matrix.columns[y]) for x, y in zip(*high_corr_var)]

# Exibir pares de colunas altamente correlacionadas
print("Pares de colunas com correlação acima de 0.95:")
for pair in high_corr_pairs:
    print(pair)

# Identificar colunas a serem removidas
columns_to_remove = set()
for pair in high_corr_pairs:
    col1, col2 = pair
    # Mantém a primeira coluna e remove a segunda
    columns_to_remove.add(col2)


# Remover colunas
df_cleaned = df.drop(columns=columns_to_remove)

# Exibir o número de colunas antes e depois da remoção

print(f"Número de colunas após a remoção: {len(df_cleaned.columns)}")

# Exibir as primeiras linhas do DataFrame após a remoção
print("\nPrimeiras linhas do DataFrame após a remoção de colunas:")
print(df_cleaned.head())

df = df_cleaned
```

## Dispersão (Histogramas e Box Plots)

Histogramas são usados para visualizar a distribuição de uma variável contínua, mostrando a frequência de ocorrência dos valores em intervalos específicos, chamados de bins.

Box plots (ou gráficos de caixa) são usados para visualizar a distribuição estatística de uma variável, mostrando a mediana, quartis e possíveis outliers.

```
# Histograma
# Análise de sobrevivência por sexo
df.groupby('sex')['survived'].mean().plot(kind='bar')
plt.title('Taxa de Sobrevivência por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Taxa de Sobrevivência')
plt.show()


# Análise de sobrevivência por classe
df.groupby('pclass')['survived'].mean().plot(kind='bar')
plt.title('Taxa de Sobrevivência por Classe')
plt.xlabel('Classe')
plt.ylabel('Taxa de Sobrevivência')
plt.show()

# Boxplots
sns.boxplot(x='survived', y='fare', data=df)
plt.title('Boxplot do Fare por Sobrevivência')
plt.xlabel('Sobreviveu')
plt.ylabel('Fare')
plt.show()
```

## Visualização MultiVariada (Multidimensionais)

t-SNE (t-Distributed Stochastic Neighbor Embedding) é uma técnica de redução de dimensionalidade usada para visualizar dados em alta dimensão em um espaço de menor dimensão, geralmente 2D ou 3D. É amplamente utilizado para análise exploratória de dados e visualização de clusters.

```
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import plotly.express as px


features = df.drop('survived', axis=1)
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Aplicar t-SNE com 3 componentes
tsne = TSNE(n_components=3, random_state=42)
tsne_results = tsne.fit_transform(features_scaled)

# Criar um DataFrame com os resultados do t-SNE
tsne_df = pd.DataFrame(tsne_results, columns=['TSNE1', 'TSNE2', 'TSNE3'])


# Visualizar os resultados do t-SNE em 3D usando Plotly
fig = px.scatter_3d(
    tsne_df,
    x='TSNE1',
    y='TSNE2',
    z='TSNE3',
    color=df['survived'],
    title='Visualização t-SNE em 3D do Conjunto de Dados TITANIC',    
)
fig.show()
```
